## Testing open set classification accuracy

```python
import numpy as np
from tqdm import tqdm
from sklearn.metrics import classification_report
```

```python
labels = ['mercedes', 'non-mercedes', 'unknown']
```

```python
list_of_preds = []
list_of_trues = []
for i,cl in enumerate(labels):
    photos = os.listdir(os.path.join("test-set", cl))
    preds = np.zeros((220))
    trues = np.zeros((220))
    list_of_images = []
    for image in photos:
        tester = image[-5:].upper()
        if not ("JP" in tester or "PNG" in tester):
            continue
        img_path = os.path.join("test-set", cl, image)
        list_of_images += [img_path]
        
    assert(len(list_of_images) == 220)
    for j,img_path in tqdm(enumerate(list_of_images)):
        preds[j] = osi.test_classification(img_path)
        trues[j] = i
    list_of_preds += [preds]
    list_of_trues += [trues]

true = np.concatenate(list_of_trues)
pred = np.concatenate(list_of_preds)
```

```python
%store true
```

```python
%store pred
```

### 3 class open set classification

```python
print(classification_report(true, pred, target_names=labels))
```

### 2 class open set classification

```python
true_2 = (true > 0) * 1
pred_2 = (pred > 0) * 1
```

```python
print(classification_report(true_2, pred_2, target_names=['mercedes', 'else']))
```
